# Processing data, correct data format, standardize and normalize data

```python
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
```

## Read cars.csv into dataframe

```python
df = pd.read_csv('./cars.csv')
df
```

## Create headers and assign to the dataframe
### Display first 5 rows

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]

df.columns=headers
df.head()
```

### Convert ? to NaN

```python
df.replace('?', np.nan,inplace=True)
df.head()
```

### Identify null values
* **True:** value is missing
* **False:** valus is not missing 

```python
missing_values = df.isnull()
missing_values.head()
```

### Specify columns with missing values

```python
for col in missing_values.columns.values.tolist():
    sum = missing_values[col].eq(True).sum()
    if sum != 0:
        print(f'{col}:{sum}') 
    
```

### Handle missing data
* Replace by mean: normalized-losses, stroke, bore, horsepower, peak-rpm
* Drop the whole row: price
* Replace by frequency: num-of-doors, replace with 4, 84% has four doors

```python
avg_normalized_loss = df['normalized-losses'].astype(float).mean(axis=0)
df.replace({'normalized-losses':np.NaN}, {'normalized-losses':avg_normalized_loss}, inplace=True)

avg_bore = df['bore'].astype(float).mean(axis=0)
df.replace({'bore':np.NaN}, {'bore':avg_bore}, inplace=True)

avg_stroke = df['stroke'].astype(float).mean(axis=0)
df.replace({'stroke':np.NaN}, {'stroke':avg_stroke}, inplace=True)

avg_horsepower = df['horsepower'].astype(float).mean(axis=0)
df.replace({'horsepower':np.NaN}, {'horsepower':avg_horsepower},inplace=True)

avg_peak_rpm = df['peak-rpm'].astype(float).mean(axis=0)
df.replace({'peak-rpm':np.NaN}, {'peak-rpm':avg_peak_rpm},inplace=True)

df.dropna(subset=['price'],axis=0,inplace=True)

df.reset_index(drop=True, inplace=True)
df.replace({'num-of-doors':np.nan},{'num-of-doors':4}, inplace=True)
```

### Correct the data format
* Get data format

```python
df.dtypes
```

```python
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```
