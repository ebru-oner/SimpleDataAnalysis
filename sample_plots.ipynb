# Data visualization samples 

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Line plot
* Expects two same size arrays as input.
* Useful; 
    * to show trends over time
    * compare multiple data series
    * to show relationship between two variables especially when one is a time series

```python
data = {'date': pd.date_range(start='1/1/2020', periods=100),
        'value': np.random.rand(100).cumsum()}
df = pd.DataFrame(data)
df.set_index('date', inplace=True)
df.plot()
plt.title('Value over time')
plt.xlabel('date')
plt.ylabel('value')
plt.show()
```

## Scatter plot
* Displays values for two variables for a set of data.
* The values are plotted as points on a graph.
* The points are plotted on a Cartesian plane.
* The horizontal axis is the independent variable.
* The vertical axis is the dependent variable.
* The points are connected by a line.
* The line is called a _scatter plot_.
* The line is not a trend/regression line.


```python
data = {'A': np.random.rand(100), 'B': np.random.rand(100)}
df = pd.DataFrame(data)

df.plot.scatter(x='A', y='B')
plt.title('A vs B')
plt.xlabel('A')
plt.ylabel('B')
plt.show()
```

## Histogram plot
* Visualize data in categorical form
* Categorization represents the frequency of data.

```python
data = { 'A': np.random.rand(1000)}
df = pd.DataFrame(data)
df['A'].plot.hist(bins=50, alpha=0.7)
plt.title('Histogram of A')
plt.xlabel('value')
plt.ylabel('frequency')
plt.show()
```

## Bar plot
* Visualizing categorical data, average, mean, median, frequency, count...
* y-axis shows the average value of the data points of a category
* x-axis shows the number of elements in each category

```python
data = {'Category':['a','b','d','r'], 'Value':[12,45,23,77]}
df = pd.DataFrame(data)
df.plot.bar(x='Category', y='Value', color='lightgreen')
plt.title('Values by category')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()   
```

## Heatmap/Pseudo color plot
* Visualize 3-dim data on a 2-dim surface
* Usually used to represent a third value

```python
x = np.linspace(0, 100, 100)
y = np.linspace(0, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

plt.imshow(Z, extent=[0,10,0,5], origin = 'lower', cmap = 'coolwarm')
plt.colorbar(label='value')
plt.title('heatmap')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

## Regression plot
* Scatter plot + fitted regression line


```python
data = {'X':[1,2,3,4,5,6,7,8], 'Y':[3,4,5,7,8,9,2,8]}
df = pd.DataFrame(data)
sns.regplot(x='X', y='Y', data=df)
plt.title('Simple linear regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

## Box and Whisker plot
* Visualizes a graphical summary of the distribution of a data set
* Shows the median, quartiles, and extremes of a data set

```python
data = {'Category': ['A','A','B','B','B','C','C','C','C','C'], 'Value':[230,45,34,67,45,34,67,99,123,87]}
df = pd.DataFrame(data)

sns.boxplot(x='Category', y='Value', data=df)
plt.title('value by category')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```
